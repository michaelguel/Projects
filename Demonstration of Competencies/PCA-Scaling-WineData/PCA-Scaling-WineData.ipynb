```python
# grabbing wine data from url source

import pandas as pd

data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', header = None)

data.columns = ['class','alcohol','malic_acid','ash','alcalinity_of_ash','magnesium','total_phenols','flavanoids','nonflavanoid_phenols','proanthocyanins','color_intensity','hue','od280/od315_of_diluted_wines','proline']

data.head()

#from sklearn.datasets import load_wine

#load_wine

# let's see the different statistical summaries for each feature

from pandas.api.types import is_numeric_dtype

for col in data.columns[1:]:
    if is_numeric_dtype(data[col]):
        print('%s:' % (col))
        print('\t Mean = %.2f' % data[col].mean())
        print('\t Standard deviation = %.2f' % data[col].std())
        print('\t Minimum = %.2f' % data[col].min())
        print('\t Maximum = %.2f' % data[col].max())

```

The feature that has the smallest values is nonflavanoid_phenols.

The feature that has the largest values is proline

The feature that has the longest range is proline

```python
# Let's look at the coorelation matrix, excluding class

corrmat = data.loc[:, data.columns != "class"].corr()

corrmat
```

```python
# Let's see which have the highest correlations

checktop = corrmat.unstack()

checktop.sort_values(ascending = False)[1:20]
```

The features that have the strongest coorelation are total_phenols and flavanoids

```python
# Let's do some PCA, first we will do it without scaling the features first

from sklearn.decomposition import PCA

# We need to seperate the features and class

seperatefeatures = data.loc[:, data.columns != "class"]

seperateclass = data.loc[:, data.columns == "class"]

# We are going to derive two PCA's

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(seperatefeatures)

principalDf = pd.DataFrame(data = principalComponents
             , columns = ['pc1', 'pc2'])

principalDf = pd.concat([principalDf,data[['class']]],axis=1)

principalDf
```

```python
# Let's graph the two PCA's according to their class, we can see that it is unscaled looking at the x and y axis

import matplotlib.pyplot as plt

colors = {1:'b', 2:'r', 3:'g'}
markerTypes = {1:'+', 2:'x', 3:'o'}

for winetype in markerTypes:
    d = principalDf[principalDf['class']==winetype]
    plt.scatter(d['pc1'],d['pc2'],c=colors[winetype],s=60,marker=markerTypes[winetype])
```

```python
# Let's do it again but scale the features before and see what changes, we can see that scaling is very important to this

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

seperatefeatures = data.loc[:, data.columns != "class"]

seperatefeatures = scaler.fit_transform(seperatefeatures)

seperateclass = data.loc[:, data.columns == "class"]

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(seperatefeatures)

principalDf = pd.DataFrame(data = principalComponents
             , columns = ['pc1', 'pc2'])

principalDf = pd.concat([principalDf,data[['class']]],axis=1)

principalDf

colors = {1:'b', 2:'r', 3:'g'}
markerTypes = {1:'+', 2:'x', 3:'o'}

for winetype in markerTypes:
    d = principalDf[principalDf['class']==winetype]
    plt.scatter(d['pc1'],d['pc2'],c=colors[winetype],s=60,marker=markerTypes[winetype])
```
