```python
import urllib
import matplotlib.pylab as plt
from bs4 import BeautifulSoup
import bs4
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import numpy as np
from yellowbrick.regressor import ResidualsPlot
import urllib
import ssl
```

```python


def get_links(url):
    with urllib.request.urlopen(url, context=ssl.SSLContext(), timeout=5) as response:
        soup = BeautifulSoup(response.read())
        meta_data = response.info()
        response_code = response.getcode()
    

    all_a_tags_with_href = set(filter(lambda x: 'href' in x.attrs, soup.body.find_all('a')))

    links_in_page = set(map(lambda x: x.attrs['href'], all_a_tags_with_href))

    links_in_page = set(map(lambda x: urllib.parse.urljoin(url, x), links_in_page))

    links_in_wikipedia = set(filter(lambda x: urllib.parse.urlparse(x).netloc.endswith('en.wikipedia.org'), links_in_page))

    return(links_in_wikipedia)

def get_10_unique(links_in_wikipedia,src_url):
    source_url = []
    webpages_on_wikipedia = set()
    pages_visited = set()

    for wiki_url in links_in_wikipedia:
        try:

        # Something could go wrong (e.g. 404 not found), so use a try
            with urllib.request.urlopen(wiki_url, context=ssl.SSLContext(), timeout=5) as response:
                content_type = response.info()['Content-Type']
            # content_type = response.get_headers().get_content_type() 
                if 'text/html' in content_type and len(webpages_on_wikipedia) < 10 and wiki_url not in pages_visited_use :
                    webpages_on_wikipedia.add(wiki_url)
                    pages_visited.add(wiki_url)
                    source_url.append(src_url)
        except:
        # Catch exception, print message, continue loop
            print(f'Unable to open {wiki_url}')
            
    return(webpages_on_wikipedia,pages_visited,source_url)

def update_url(r,urls):
    new_url = ''
    for y in r:
        if y not in urls and '/wiki/Wikipedia:' not in y and '/wiki/Category:' not in y and '/wiki/File:' not in y and '/wiki/Special:' not in y:
            new_url = y
            break
    return(new_url)

def get_titles(pages_visited_use):
    page_titles = []
    for wiki_url in pages_visited_use:
        try:
            with urllib.request.urlopen(wiki_url, context=ssl.SSLContext(), timeout=5) as response:
                soup = BeautifulSoup(response.read())
            title = soup.title.text
            page_titles.append(title)
        except:
        # Catch exception, print message, continue loop
            print(f'Unable to open {wiki_url}')
            page_titles.append('Unable to open and get title')
    return(page_titles)

    

    
```

```python
url = 'https://en.wikipedia.org/wiki/Main_Page'

count = 0

urls = set()

webpages_on_wikipedia_use = set()

pages_visited_use = set()

source_url_use = []


while count <= 9:

    urls.add(url)

    print(url)

    links_in_wikipedia = get_links(url)

    update_webpages,update_pages,update_src_url = get_10_unique(links_in_wikipedia,url)

    source_url_use.extend(update_src_url)

    webpages_on_wikipedia_use.update(update_webpages)

    pages_visited_use.update(update_pages)


    count += 1
    print(count)
   
    r = list(update_webpages)
    i = update_url(r,urls)
    url = i 
```
